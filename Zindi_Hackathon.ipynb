### We have our dataset fit for training , so we go into modelling and using GridSearchCV to find the best algorithm

```python
from sklearn.model_selection import train_test_split
x_train ,x_test , y_train ,y_test =train_test_split(x_scaled , y , random_state = 10 ,test_size = 0.3)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
len(y_train)
```

```python
len(y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import GridSearchCV
```

```python
model_parameters = {
    'lr':{
        'model' : LogisticRegression(),
        'params':{
            'C' : [2,5 ,10],
            'max_iter' :[100 ,500 ]
        }
    },
    'dt':{
        'model':DecisionTreeClassifier(),
        'params':{
            'criterion' :['gini' ,'entropy'],
            'max_depth' : [5 ,7 ,10],
            'min_samples_split' : [2 ,5 ,10],
            'min_samples_leaf' :[1 ,2 ,5]
        }
    },
    'rf':{
        'model' : RandomForestClassifier(),
        'params' :{
            'n_estimators' : [50 ,100],
            'min_samples_split' : [2 ,5 ,10],
            'min_samples_leaf' :[1 ,2 ,5],
        }
    },
    'svm':{
        'model' :SVC(),
        'params' : {
            'C' :[2,5,10],
            'kernel' :['linear' ,'rbf'],
        }
    },
    'knn':{
        'model':KNeighborsClassifier(),
        'params' :{
            'n_neighbors' :[1 ,3 ,5 ,7 ]
            
        }
    },
    'nb':{
        'model':GaussianNB(),
        'params' :{
            'priors': [[0.5, 0.5], [0.2, 0.8], [0.8, 0.2]] ,
        }
    }
}
```

```python
from sklearn.model_selection import KFold ,StratifiedKFold
kf = KFold(n_splits = 5 ,shuffle = True ,random_state = 42)
```

```python
scores = []
for model_name , mp in model_parameters.items():
    clf = GridSearchCV(mp['model'] ,mp['params'] ,return_train_score = False ,cv  = kf)
    clf.fit(x_train ,y_train)
    scores.append({
        'model':model_name,
        'best_score' : clf.best_score_,
        'best_params' : clf.best_params_
    })
```

```python
df_scores  = pd.DataFrame(scores)
df_scores 
```

```python
plt.bar(df_scores['model'] ,df_scores['best_score'] ,color = ['red' ,'blue' ,'green' ,'orange' ,'purple' ,'black'])
plt.title("Checking the accuracy of Classification Algorithms")
plt.xlabel("ML Algorithms")
plt.ylabel("Accuracy")
plt.show()
```

### We have seen that Random Forest Algorithm have the highest accuracy score , so we will be using the algorithm to train our model that will be used for predictions

```python
model = RandomForestClassifier(min_samples_leaf = 5 ,min_samples_split = 2 ,n_estimators = 100)
model.fit(x_train ,y_train)
```

```python
model.score(x_test ,y_test)
```
